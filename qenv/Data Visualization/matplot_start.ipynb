```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), linestyle='solid')
plt.plot(x, np.tan(x), linestyle='dashed');
```

<p> استفاده از نقطه‌ویرگول (;) در انتهای خطوط

شاید از خود بپرسید دلیل وجود نقطه‌ویرگول (semicolon) در انتهای خطوط کد بالا چیست؟ مگر قرار نبود پایتون نقطه‌ویرگول نداشته باشد؟ در قسمت زیر به این سوال پاسخ داده‌ایم.

در نت‌بوک، هنگامی‌که شما اسم متغیر را در سلول می‌نوشتید، مقدار متغیر در خروجی نمایش داده می‌شد. یعنی اگر متغیر ما لیست پایتون، آرایه‌ی نامپای، رشته یا انواع دیگر متغیر باشد، با نوشتن اسم آن در سلول و اجرای آن، مقدار آن متغیر در خروجی نمایش داده می‌شود.

این ویژگی خوبی است که ما هم تا اینجا بارها از آن استفاده کرده‌ایم. اما اگر موقعیتی پیش بیاید که ما نیاز به دیدن مقدار متغیر نداشته باشیم، چه باید کرد؟ در این حالات با قرار دادن سمی‌کالن در انتهای آن خط، مقدار آن متغیر در خروجی نمایش داده نمی‌شود.

برای نمایش نمودارها در این فصل، ما «شیء شکل» (در ادامه با آن آشنا خواهید شد) را در انتهای سلول می‌نویسیم و سلول را اجرا می‌کنیم. اگر سمی‌کالن در انتهای خط نگذاریم، علاوه بر عکس نمودار، محتوای شیء شکل نیز در خروجی نمایش داده می‌شود. چاپ مقدار شیء شکل ضرری برای کسی ندارد و ما برای زیبا‌تر شدن خروجی‌ها، آن را نمایش نمی‌دهیم.</p>

<p>عناصر شکل و صفحه‌‌محور

عناصر شکل (figure) و صفحه‌محورها (axes)، دو عنصر اساسی طرح‌ها (plots) در مت‌پلات هستند. شکل یا همان figure عنصری است که بالاترین سطح را دارد و سایر عناصر یک طرح (plot) را در بر می‌گیرد. به این ترتیب وقتی می‌خواهیم نموداری رسم کنیم، اولین چیزی که ساخته می‌شود یک figure است که عناصر دیگر داخل آن قرار می‌گیرند. اگر ما این عنصر را نسازیم، خود پایتون به‌صورت پیش‌فرض آن را می‌سازد. همچنین ما می‌توانیم به‌صورت زیر آن را بسازیم:</p>

```python
fig = plt.figure()
```

<p>عنصر دیگری که در رسم کردن نمودارها نقش کلیدی دارد، صفحه‌محور است. هر نموداری که رسم می‌کنیم داخل یک صفحه‌محور قرار می‌گیرد. صفحه‌محور عناصری مانند محورها، صفحه‌ی مختصات، عنوان نمودار و سایر جزئیات آن را نشان می‌دهد. در شکل بالا، یک صفحه‌محور داشتیم که نمودار را در همان رسم کردیم. اما می‌توان شکل‌هایی شامل چند صفحه‌محور هم ساخت. برای این منظور می‌توانیم یک صفحه‌محور را در یک مختصات خاص از شکل بسازیم و یا شکل را به جدولی از صفحه‌محورها تقسیم کنیم. در ادامه به هر دو روش خواهیم پرداخت. پیش از آن لازم است کمی درباره دو نوع شیوه کار با مت‌پلات بدانیم. در تصویر زیر، یک شکل که شامل دو صفحه‌محور است را مشاهده می‌کنید.</p>

<p>یکی از مواردی که ممکن است کاربران را در کار با مت‌پلات دچار ابهام کند، پشتیبانی آن از دو نوع رابط مختلف است. رابط اول، یک رابط وضعیت‌مند (state-based) است که کار با آن مشابه کار با متلب است و رابط دوم یک رابط شی‌ءگراست که توانایی بیشتری دارد.</p>

<h3>رابط متلبی</h3>
<p>از آنجایی که مت‌پلات در ابتدا به عنوان جایگزین پایتونی متلب نوشته شد، بسیاری از دستورات آن از نظر نحوی مشابه متلب تعریف شده‌اند. اصلی‌ترین رابط برای استفاده از این دستورات متلبی رابط pyplot است. شیوه‌ی کار این رابط به شکل وضعیت‌مند است: به این معنی که همواره شکل و صفحه‌محور اخیر را نگه می‌دارد و هر زمان که کاربر یک تابع را فراخوانی کند، تغییرات روی شکل و صفحه‌ی ذخیره‌شده اعمال می‌شود. در کدی که در ابتدا نوشتیم یک شکل ابتدایی و یک صفحه‌محور داخل آن به‌صورت غیر‌مستقیم تعریف شده‌اند و با صدا زدن تابع plot نمودارهای مربوطه روی آن رسم می‌شوند.

با صدا زدن تابع plt.axes می‌توانیم یک صفحه‌محور جدید بسازیم و مختصات نقطه‌ی پایین-چپ، طول و عرض آن را نسبت به اندازه‌ی شکل در قالب یک لیست ورودی دهیم (برای فهم بهتر به مثال زیر و اعداد نسبی دقت کنید). پس از تعریف، این صفحه‌محور به عنوان صفحه‌محور اخیر در نظر گرفته می‌شود و تا زمانی که با صفحه‌محور دیگری جایگزین نشود، توابع مختص به یک صفحه‌محور رابط روی آن اعمال می‌شوند. در کد زیر، یک نمونه استفاده از آن را مشاهده می‌کنید.</p>

```python
plt.figure() # یک شکل می‌سازد
plt.axes() # یک صفحه محور استاندارد داخل شکل اخیر می‌سازد
plt.plot(x, np.exp(x))
plt.axes([0.2, 0.4, 0.5, 0.3]) # در لیست مختصات نقطه پایین چپ و طول و عرض صفحه محور نسبت به ابعاد شکل قرار گرفته‌اند
plt.plot(x, np.log(x), color='red', linestyle='dashed')
```

<p>همان‌طور که پیش‌تر گفتیم، صفحه‌محورها را می‌توان به شکل یک جدول هم داخل شکل قرار داد. برای این منظور می‌توانیم از تابع plt.subplot استفاده کنیم. به این تابع تعداد سطرها و ستون‌های جدولِ صفحه‌محورها، و سپس شماره‌ی صفحه‌محور داخل جدول را می‌دهیم تا ساخته شود (شماره‌ها به‌صورت ستونی و سپس سطری از بالا چپ جدول شروع می‌شوند و مشابه متلب شماره نخستین صفحه‌محور برابر یک است). توجه داشته باشید که هر صفحه‌محور پس از ساخت به عنوان صفحه‌محور اخیر در نظر گرفته می‌شود. کد زیر، نمونه استفاده از این روش را نشان می‌دهد.</p>

```python
plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x))
plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x), linestyle='dashdot', color='red')
plt.subplot(2, 2, 3)
plt.plot(x, np.exp(x), color='green', linestyle='dashed')
plt.subplot(2, 2, 4)
plt.plot(x, x**2, color='magenta')
```

<h3>رابط شی‌ءگرا</h3>
<p>استفاده از رابط شیءگرا به این ترتیب است که در ابتدا به کمک رابط pyplot یک شکل شامل چند صفحه‌محور می‌سازیم و سپس، عملیات‌ها را با صدا زدن توابع خود شکل و صفحه‌محورها انجام می‌دهیم. به این ترتیب می‌توانیم هر موقع که خواستیم با صدا زدن تابع یکی از صفحه‌محورهای موجود، نموداری داخل آن رسم کنیم.

با صدا زدن تابع add_axes شکل و دادن مختصات و طول و عرض می‌توانیم یک صفحه‌محور داخل شکل بسازیم. بنابراین می‌توانیم معادل نمودار نخستی که با رابط متلبی ساختیم را به شکل زیر با رابط شیءگرا بسازیم.</p>

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = fig.add_axes([0.2, 0.4, 0.5, 0.3])
ax1.plot(x, np.exp(x))
ax2.plot(x, np.log(x), color='red', linestyle='dashed')
```

<p>همچنین با استفاده از تابع add_subplot شکل، مشابه روش متلبی می‌توانیم صفحه‌محورها را داخل یک جدول بسازیم و سپس نمودارهای هر یک را رسم کنیم تا نمودار معادل دومین نمودار رسم شده در بالا را به دست آوریم.</p>

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)
ax1.plot(x, np.sin(x))
ax2.plot(x, np.cos(x), color='red', linestyle='dashdot')
ax3.plot(x, np.exp(x), color='green', linestyle='dashed')
ax4.plot(x, x**2, color='magenta', linestyle='solid')
```

<p>برای این منظور، راه ساده‌تری هم وجود دارد و آن استفاده از تابع subplots در رابط pyplot است. این تابع یک شکل و یک آرایه از صفحه‌محورهای داخل آن با ابعادی که ورودی گرفته، می‌سازد. به این ترتیب می‌توانیم با استفاده از آرایه به صفحه‌محورها دسترسی داشته باشیم.</p>

```python
fig, ax = plt.subplots(2, 3, sharex='col', sharey='row')
```
