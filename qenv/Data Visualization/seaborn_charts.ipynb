<p>در این قسمت، بعضی از نمودار‌هایی که در تحلیل داده می‌توانند استفاده شوند و توسط سیبورن قابل رسم هستند را بررسی می‌کنیم.</p>
<h3>کار با regplot</h3>
<p>


نمودار پراکندگی که از آن برای بررسی ارتباط دو متغیر استفاده می‌شد را به‌خاطر دارید؟ هنگام کار با نمودار پراکندگی، خودمان باید مسیر و روند داده را به شکل ذهنی مجسم می‌کردیم، اما به کمک سیبورن می‌توان با خطی این روند را نشان داد.

در درسنامه‌ی نمودار پراکندگی، رابطه بین قیراط الماس و قیمت آن‌ را نمایش دادیم. حالا این کار را با تابع regplot سیبورن انجام می‌دهیم تا روند موجود نیز نمایش داده شود.</p>

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()

fig , ax = plt.subplots()

price = [921, 3087, 7196, 4201, 982, 3534, 9930, 8518, 12825, 7404, 4032, 878, 1240, 945, 10446, 5530, 720, 2846, 3064]
carat = [0.42, 0.76, 1.21, 0.91, 0.4, 1.02, 1.24, 1.5, 1.31, 1.13, 0.57, 0.41, 0.49, 0.42, 1.14, 0.91, 0.32, 0.77, 0.8]

sns.regplot(price, carat , ax= ax)
ax.set_title("carat vs price for diamonds")
ax.set_xlabel("carat")
ax.set_ylabel("price")
```

<p>بر خلاف نموداری که مت‌پلات کشیده بود و فقط نقاط را نشان می‌داد، با سیبورن ما اطلاعات بیشتری در اختیار داریم. خط مسیر داده را نشان می‌دهد و حاشیه‌ی کم‌رنگ آن بر اساس میزان اطمینان به خط کشیده‌شده کوچک و بزرگ می‌شود. در واقع سیبورن با استفاده از داده‌ها یک مدل رگرسیونی می‌سازد که روند داده را پیش‌بینی می‌کند.

این تابع، در واقع یک رگرسیون خطی روی داده انجام می‌دهد که البته با تنظیم پارامترها می‌توان رگرسیون غیرخطی هم روی داده انجام داد. از شما انتظار نمی‌رود که در این دوره با این مفاهیم کاملاً آشنا باشید. رگرسیون انواع مختلف و پارامترهای زیادی دارد. تعدادی از پارامترهای رگرسیون، جزو پارامترهای ورودی تابع regplot هستند که در ادامه به آن‌ها می‌پردازیم. برای کسب اطلاعات بیشتر در مورد رگرسیون می‌توانید صفحه‌ی ویکی‌پدیای آن را از اینجا ببینید.</p>

<h3>کار با displot</h3>
<p>

یکی دیگر از توابع مهم سیبورن که بسیار کاربردی هم می‌باشد، تابع displot می‌باشد. تابع hist مت‌پلات برای ما هیستوگرام داده را می‌کشید که در آن داده‌ها در سطل‌هایی تقسیم می‌شدند و سپس تعداد داده موجود در هرسطل به شکل نموداری میله‌ای به نمایش در می‌آمد.

سپس دیدیم که برای نمایش تابع چگالی داده، باید از واسط پانداز کمک بگیریم و خروجی آن خطی پیوسته بود که توزیع داده را نمایش می‌داد. حال سیبورن با تابع displot همزمان چگالی و هیستوگرام را برای ما تولید می‌کند. (در حالت پیش‌فرض، محور عمودی، تعداد عناصر هر سطل را نشان می‌دهد و با تنظیم پارامتر‌ stat که توضیحات آن در ادامه آمده، محور عمودی می‌تواند احتمال یا درصد را نشان دهد.)

در این مثال و مثال‌های پیش‌رو در این درسنامه بار دیگر از اطلاعات بازیکنان فوتبال که از بازی FIFA 21 به‌دست آمده است، استفاده می‌کنیم.</p>

```python
import pandas as pd

df = pd.read_csv('fifa21.csv')
sns.displot(df, x='height', bins = 30, color = 'red',kde=True);
```

<h3>کار با jointplot</h3>
<p>

یکی دیگر از نوآوری‌های سیبورن، jointplot‌ می‌باشد. این نمودار از ۳ بخش تشکیل شده است. دو بخش آن از یک جنس هستند و می‌توانند هیستوگرام یا تابع چگالی یک متغیر باشند. بخش سوم وضعیت دو متغیر را نسبت به هم نشان می‌دهد که می‌تواند نمودار پراکندگی، تابع چگالی یا هر نمایش دیگری که بتواند وضعیت دو متغیر را نسبت به هم نشان دهد، باشد. (برای آشنایی با انواع دیگر نمایش، می‌توانید این لینک را ببینید.) در مثال زیر، بار دیگر وضعیت وزن و قد بازیکنان را بررسی می‌کنیم.</p>

```python
plot = sns.jointplot(data = df, x='height' ,y='weight', kind = 'kde', color='purple')
```

<p>
از آن‌جایی که سیبورن بیشتر برای نمودارهای آماری مناسب است، در اکثر توابع آن یک پارامتر data وجود دارد که می‌توانیم فایل دیتافریم یا یک آرایه چند‌بعدی نامپای را به آن بدهیم و با پارامترهای دیگر (در اینجا x و y) ستون‌هایی از داده که می‌خواهیم رسم شوند را مشخص کنیم.

ستون‌های قد و وزن را انتخاب کرده‌ایم و با مقداردهی پارامتر kind به kde، نمودار چگالی ستون‌های قد و وزن رسم شده است. در مربع بزرگ‌تر وضعیت این دو ستون نسبت به هم به شکل یک طرح دو بعدی نمایش داده شده و قسمت‌هایی از شکل که خطوط به هم نزدیک‌تر هستند (قد حدود ۱۸۰، وزن حدود ۷۵)، نشان‌دهنده‌ی تجمع داده می‌باشد. یعنی بازیکنانی که قد و وزن‌شان در آن ناحیه است، تعدادشان بیشتر از بقیه است.

این نمودار از سه بخش تشکیل شده‌ است و نمی‌توان آن را بر روی یک صفحه‌محور مت‌پلات کشید. هر بار که از این تابع استفاده می‌کنیم، یک شکل جدید ایجاد می‌شود که بر خلاف نمودارهای قبل، قابل شخصی‌سازی با توابع مت‌پلات نیست.</p>
