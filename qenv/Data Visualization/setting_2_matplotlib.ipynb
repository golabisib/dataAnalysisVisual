<h3>مارکرها</h3>
<p>

با تابع plot که برای رسم نقطه یا خط استفاده می‌شود در درسنامه‌ی اول تنظیمات آشنا شدیم. همچنین تابع پراکندگی را با هم دیدیم که وضعیت و هم‌بستگی دو متغیر نسبت به هم را نشان می‌دهد. در مواردی لازم می‌شود که نقاط تشکیل‌دهنده‌‌ی خط را متمایز کنیم و یا با شکل‌هایی تمرکز بر آن‌ها ‌را بیشتر کنیم، با استفاده از مارکر (marker) می‌توان شکل و اندازه‌ی نقاط را تغییر داد. به مثال زیر توجه کنید:</p>

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()

x = np.array([1,2,3,4,5])
y = x * 2

ax.plot(x + 2 ,y,marker = 'x')                      # blue
ax.plot(x + 1 ,y,marker = 'o')                      # orange
ax.plot(x ,y ,marker = 'p')                         # green
ax.plot(x - 1 ,y,marker = '*', markersize = 5)      # red
ax.plot(x - 2 ,y,marker = '1', markersize = 10)     # purple
ax.plot(x - 3 ,y,marker = 'v', markersize = 15)     # brown
ax.plot(x - 4 ,y,marker = 'h', markersize = 25);    # pink
```

<p>توضیح تصویر

در شکل بالا، یک خط (شامل ۵ نقطه) را با مارکرهای گوناگون رسم کرده‌ایم. علاوه بر مارکرهایی که در شکل بالا می‌بینید، مارکر‌های دیگری هم وجود دارد که می‌توانید از </p>
<a href='https://matplotlib.org/stable/api/markers_api.html'>این لینک‌ </a>
<p>در مورد آن‌ها بیشتر بدانید.

علاوه بر نوع مارکرها، اندازه‌ها هم تغییر یافته‌اند که این کار به کمک پارامتر markersize انجام گرفته است.

در تابع scatter هم به همین منوال عمل می‌کنیم، کد زیر را ببینید:</p>

```python
import matplotlib.pyplot as plt
import numpy as np
fig , ax = plt.subplots()

ax.scatter(np.random.uniform(-4,4,7),np.random.randint(-4,4,7),marker = 'p') # blue
ax.scatter(np.random.uniform(-4,4,7),np.random.randint(-4,4,7),marker = 'x') # orange
ax.scatter(np.random.uniform(-4,4,7),np.random.randint(-4,4,7),marker = 'o') # green
ax.scatter(np.random.uniform(-4,4,7),np.random.randint(-4,4,7),marker = '*') # red
ax.scatter(np.random.uniform(-4,4,7),np.random.randint(-4,4,7),marker = '1') # purple
ax.scatter(np.random.uniform(-4,4,7),np.random.randint(-4,4,7),marker = 'v') # brown
ax.scatter(np.random.uniform(-4,4,7),np.random.randint(-4,4,7),marker = 'h');# pink
```

<h3>انواع خط</h3>
<p>

یکی دیگر از تنظیماتی که می‌تواند خیلی مفید باشد، نوع خط می‌باشد. برای تغییر نوع خط، هنگام رسم آن در تابع plot، متغیر linestyle را تغییر می‌دهیم.

برای تغییر ضخامت خط هم باید متغیر linewidth را تغییر دهید:</p>

```python
import matplotlib.pyplot as plt
import numpy as np
fig , ax = plt.subplots()

x = np.array([1,2,3,4,5])
y = np.array([1, 1, 1, 1, 1])
ax.plot(x ,y - 1 ,linestyle = 'solid')
ax.plot(x ,y - 2 ,linestyle = 'dotted')
ax.plot(x ,y - 3 ,linestyle = 'dashed')
ax.plot(x ,y - 4 ,linestyle = 'dashdot')
ax.plot(x ,y - 5 ,linestyle = 'dashdot', linewidth = 4)
ax.plot(x ,y - 6 ,linestyle = 'dashdot', linewidth = 8);
```

<h3>کار با رنگ‌ها</h3>
<p>

در مت‌پلات می‌توان رنگ بسیاری از عناصر را تغییر داد. برای مثال، رنگ عنوان نمودار، رنگ خطوط، رنگ محورها، رنگ زمینه نمودار و غیره.

در این قسمت رنگ‌های مختلف را امتحان می‌کنیم:</p>

```python
import matplotlib.pyplot as plt
import numpy as np
fig , ax = plt.subplots()

x = np.array([1,2,3,4,5])
y = np.array([1, 1, 1, 1, 1])
ax.plot(x ,y - 1 ,linestyle = 'solid', linewidth = 8, color = "b")
ax.plot(x ,y - 2 ,linestyle = 'solid', linewidth = 8, color = "red")
ax.plot(x ,y - 3 ,linestyle = 'solid', linewidth = 8, color = "chocolate")
ax.plot(x ,y - 4 ,linestyle = 'solid', linewidth = 8, color = "cyan")
ax.plot(x ,y - 5 ,linestyle = 'solid', linewidth = 8, color = "#eaaff9")
ax.plot(x ,y - 6 ,linestyle = 'solid', linewidth = 8, color = "tomato")
ax.set_facecolor("#aafff5")
ax.set_ylabel('axis y', color = "yellow")
ax.set_xlabel('axis x', color = "yellow")
ax.set_title("colors", color = 'lime')
fig.set_facecolor((.18, .31, .31));
```

<h3>کار با علامت محورها</h3>
<p>

```python
fig , ax = plt.subplots()
tehran_tir1400 = [31.1, 31.7, 32.5, 31.9, 31.8, 32.7, 33.1, 32.9, 32.7, 32.4, 32.2,
                  32.6, 33.5, 35.4, 35.6, 35.8, 32.7, 30.4, 30. , 31.1, 32.7, 31.5,
                  30.4, 28.6, 27.9, 30.3, 32.8, 31.9, 31.7, 32.7, 34.4]
ax.plot(tehran_tir1400)
ax.set_xticks([6, 13, 20, 27])
ax.set_xticklabels(labels=['first week', 'second week', 'third week', 'fourth week'], rotation=30, color = 'green')
ax.set_xlabel('weeks', color="blue")
ax.set_ylabel('temprature', color="blue");
ax.set_title('Tehran average temprature tir 1400', color = 'blue')
```
