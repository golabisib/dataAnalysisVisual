<p>با بسیاری از نکات ریز و درشت مت‌پلات آشنا شدیم. نمودار‌های مهم آن‌ را شناختیم و راه‌های تغییر تنظیمات نمودارها و شخصی‌سازی آن‌ها را با هم دیدیم.

شما اکنون می‌توانید نمودارهای مختلفی متناسب با نیاز و داده‌های در دسترس خود ایجاد کنید؛ اما این پایان ماجرا نیست. در ابتدای فصل گفتیم که مصور‌سازی داده، نیمی هنری و نیمی علمی است. مدت زیادی از معرفی مت‌پلات گذشته و نمودارهایی که مت‌پلات می‌کشد، ظاهری رنگ و رو رفته دارند. درست است که شما به همه چیز دسترسی دارید و می‌توانید همه چیز را تغییر دهید تا در نهایت به نمودار زیبای خود برسید، اما این کار نیازمند نوشتن کدهای تکراری بسیاری است و زمان زیادی ممکن است از شما بگیرد.

مت‌پلات با معرفی ویژگی style تا حدودی این مشکل را حل کرد. قبل از کشیدن نمودارها، با تغییر مقدار پیش‌فرض style می‌توان تم نمودارها را تغییر داد و بدون نوشتن کد اضافی به نمودار‌های زیباتری دست‌ یافت.
سیبورن چیست؟

راه‌ دیگر برای داشتن نمودار‌های زیبا استفاده از کتابخانه‌ی سیبورن (seaborn) است. سیبورن هم مانند رابط پانداز بر پایه‌ی مت‌پلات نوشته شده و در بطن خود از مت‌پلات برای کشیدن نمودارها استفاده می‌کند. سیبورن بیشتر برای رسم نمودارهای آماری استفاده می‌شود و توانایی کشیدن نمودارهایی را دارد که مت‌پلات قادر به رسم آن‌ها نیست. تعدادی از این نمودارها را در بخش بعد می‌بینیم.

با اینکه فقط با استفاده از سیبورن بسیاری از نمودارها را رسم کرد و کار را پیش برد، اما برای سفارشی‌کردن نمودارها و تغییر همه اجزای آن، شناخت مت‌پلات و واسط آن امری ضروریست. هماهنگی سیبورن با مت‌پلات باعث می‌شود تا سیبورن هم مانند مت‌پلات، در بسیاری از محیط‌ها و سیستم‌عامل‌ها قابل استفاده باشد. داشتن دانشی ترکیبی از مت‌پلات و سیبورن باعث می‌شود تا شما هم نمودارهایی زیبا داشته باشید و هم در صورت نیاز، نمودارهای پیچیده و منحصر‌به‌فردی را بتوانید ایجاد کنید.

به نمودار زیر توجه کنید که با مت‌پلات کشیده شده‌ است:</p>

```python
import matplotlib.pyplot as plt

fig , axes = plt.subplots(1,2, figsize = (9,3.5))
fig.suptitle('website analysis')

axes[0].set_title('views')
axes[0].set_xlabel('day')
axes[0].set_ylabel('count')
axes[0].plot([10,15,12,18,9,13,12],label = 'this week')
axes[0].plot([8,9,13,10,11,14,15],label = 'last week')
axes[0].legend()

axes[1].set_title('new accounts')
axes[1].set_xlabel('day')
axes[1].set_ylabel('count')
axes[1].bar([0,1,2,3,4,5,6],[6,4,4,3,2,4,5])
```

<p>می‌توان از سیبورن به عنوان تِمی برای مت‌پلات استفاده کرد. برای این کار کافیست که سیبورن را import‌ کنیم و متد set_theme آن را صدا بزنیم. پس از آن هر نموداری که با مت‌پلات کشیده می‌شود، ظاهر سیبورن را به خود می‌گیرد. در عکس زیر همان نمودار بالا را با مت‌پلات کشیده‌ایم، اما قبل از اجرای آن تابع sns.set_theme() را صدا زده‌ایم:</p>

```python
import seaborn as sns
sns.set_theme()

fig , axes = plt.subplots(1,2, figsize = (9,3.5))
fig.suptitle('website analysis')

axes[0].set_title('views')
axes[0].set_xlabel('day')
axes[0].set_ylabel('count')
axes[0].plot([10,15,12,18,9,13,12],label = 'this week')
axes[0].plot([8,9,13,10,11,14,15],label = 'last week')
axes[0].legend()

axes[1].set_title('new accounts')
axes[1].set_xlabel('day')
axes[1].set_ylabel('count')
axes[1].bar([0,1,2,3,4,5,6],[6,4,4,3,2,4,5])
```

<p>همان‌طور که می‌بینید، نمودار ظاهر مدرن‌تر و زیباتری به خود گرفته است و در نمودارهای پیچیده‌تر، زیبایی آن حتی بیشتر دیده می‌شود. راه دیگر استفاده از سیبورن، استفاده از توابع آن برای رسم نمودارها می‌باشد. در درسنامه‌ی بعدی به این موضوع می‌پردازیم.</p>
